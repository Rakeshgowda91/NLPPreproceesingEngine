```python
import re
import string
import nltk
import emoji

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
def remove_urls(text):
    return re.sub(r'http\S+|www\S+', '', text)

def remove_emails(text):
    return re.sub(r'\S+@\S+', '', text)

def remove_emojis(text):
    return emoji.replace_emoji(text, replace='')

def clean_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

def remove_whitespace(text):
    return re.sub(r'\s+', ' ', text).strip()

def tokenize_text(text):
    return word_tokenize(text)

def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

def apply_stemming(tokens):
    return [stemmer.stem(word) for word in tokens]

def apply_lemmatization(tokens):
    return [lemmatizer.lemmatize(word) for word in tokens]
```

```python
def preprocess_text(text):
    text = remove_urls(text)
    text = remove_emails(text)
    text = remove_emojis(text)

    text = clean_text(text)
    text = remove_whitespace(text)

    tokens = tokenize_text(text)
    tokens_no_stop = remove_stopwords(tokens)

    stemmed = apply_stemming(tokens_no_stop)
    lemmatized = apply_lemmatization(tokens_no_stop)

    return {
        "clean_text": text,
        "tokens": tokens,
        "no_stopwords": tokens_no_stop,
        "stemmed": stemmed,
        "lemmatized": lemmatized
    }
```

```python
sample_texts = [
    "Hello!!! Visit https://example.com 😊",
    "Contact me at test@gmail.com for details.",
    "This is a SIMPLE text with numbers 12345 and symbols!!!",
    "   Extra    spaces    should be handled properly.   ",
    "NLP is amazing!!! 🚀🔥 #AI"
]
```

```python
for i, text in enumerate(sample_texts):
    print(f"\n--- Sample {i+1} ---")
    print("Original:", text)

    result = preprocess_text(text)

    print("Clean Text:", result["clean_text"])
    print("Tokens:", result["tokens"])
    print("No Stopwords:", result["no_stopwords"])
    print("Stemmed:", result["stemmed"])
    print("Lemmatized:", result["lemmatized"])
```
